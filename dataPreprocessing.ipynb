```python
import pandas as pd
```

Read CSV

```python
df = pd.read_csv('pollution_us_2014_2016.csv')
df
```

Delete null row

```python
df = df.dropna(axis=0, how='any')
```

Check Null still exists

```python
check_null = df.isnull().sum()
check_null
```

Delete index and units

```python
df.drop(['Unnamed: 0', 'State Code', 'County Code', 'Site Num', 'Address', 'County', 'NO2 Units', 
        'O3 Units', 'SO2 Units', 'CO Units'], axis = 1, inplace= True)
column_names = df.columns
print(column_names)
```

Change Date Time format

```python
# Convert date column to datetime format
df['Date Local'] = pd.to_datetime(df['Date Local'])

# Change the date format to DD/MM/YYYY
df['Date Local'] = df['Date Local'].dt.strftime('%d/%m/%Y')
```

Analysis only Year 2015

```python
df = df[~df['Date Local'].str.contains('2014|2016')]
```

Check Duplicated

```python
duplicate_dates_count = df.duplicated()
duplicate_dates_count.value_counts()
```

Demolish duplicate rows
1. Different value in last 5column (CO)
2. Get duplicate rows to count for mean 
3. updated value in both rows
4. Remove duplicate rows

```python
# Group the DataFrame by all columns except the last five
grouped = df.groupby(df.columns[:-5].tolist())

# Calculate the mean of the last five columns for each group
mean_values = grouped[df.columns[-5:]].transform('mean')

# Update the original DataFrame with the mean values for the duplicate rows
df[df.columns[-5:]] = mean_values

# Drop the duplicate rows
df.drop_duplicates(subset=df.columns[:-5].tolist(), keep='first', inplace=True)

# Reset the index of the DataFrame
df.reset_index(drop=True, inplace=True)
```

```python
duplicate_dates_count = df.duplicated()
duplicate_dates_count.value_counts()
```

Check [City = Los Angeles] <- found duplicated rows

```python
los_angeles_cities = df[df['City'] == 'Los Angeles']
print(los_angeles_cities)
```
